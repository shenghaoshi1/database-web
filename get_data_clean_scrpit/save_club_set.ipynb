```python
import pandas as pd
import pickle
df=pd.read_csv('player_cleaned.csv',header=0)
club=df['Club']
club_set=set()
for x in club:
    if type(x)!=str :
        continue
    if x[0:3]=='1. ':
        x=x[3:]
    if x!='':
        club_set.add(x)
print(len(club_set))
fileObject = open('clubset.pkl','wb')
pickle.dump(club_set,fileObject)
fileObject.close()
club_set
```
